```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### 1

```python
arr1 = np.array([1, 2, 3, 4, 5])
print("Sum of all elements:", np.sum(arr1))
```

### 2

```python
arr2 = np.array([[5, 10, 15], [20, 25, 30], [35, 40, 45]])
print("\n3x3 Array:\n", arr2)
print("Sum of each row:", np.sum(arr2, axis=1))
print("Sum of each column:", np.sum(arr2, axis=0))
print("Second maximum element:", np.sort(arr2.flatten())[-2])
```

### 3

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print("\nMatrix Multiplication:\n", np.dot(A, B))
```

### 4

```python
data = {'X':[1,2,3,4,5], 'Y':[2,3,4,5,6], 'Z':[3,4,5,6,7]}
df1 = pd.DataFrame(data)
print("\nPowers of array values:")
print(df1)

df2 = df1.copy()
df2 = df2.apply(lambda col: np.power(col, df1[col.name]))
print("\nElement-wise Power:")
print(df2)
```

### 5

```python
examData = {
    'name': ['satyam', 'yajas', 'shubhang', 'anurag'],
    'score': [12.5, 9, 16.5, np.nan],
    'attempts': [1, 3, 2, 3],
    'qualify': ['yes', 'no', 'yes', 'no']
}
labels = ['a', 'b', 'c', 'd']
df3 = pd.DataFrame(examData, index=labels)
print("\nFirst 3 rows:")
print(df3.iloc[:3])

# 6

print("\nDataFrame before replacing missing values:\n", df3)
df3.fillna(value={"score": df3["score"].mean()}, inplace=True)
print("After replacing missing values:\n", df3)
```

### 7

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(12, 8))

# Line plot
plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.title("Line Plot")

# Scatter plot
plt.subplot(2, 2, 2)
plt.scatter(x, y, color='red')
plt.title("Scatter Plot")

# Bar plot
plt.subplot(2, 2, 3)
plt.bar(['A', 'B', 'C'], [5, 7, 3])
plt.title("Bar Plot")

# Pie chart
plt.subplot(2, 2, 4)
plt.pie([10, 20, 30], labels=['X', 'Y', 'Z'], autopct='%1.1f%%')
plt.title("Pie Chart")

plt.tight_layout()
plt.show()
```
